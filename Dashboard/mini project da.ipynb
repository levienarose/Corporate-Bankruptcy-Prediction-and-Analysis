In the corporate world, a business rarely collapses without warning. Similar to a medical professional monitoring a patient’s vital signs such as blood pressure, heart rate, and temperature to diagnose an illness, a data analyst can evaluate key financial metrics to predict whether a company is heading towards success or bankruptcy while also maintaining its status in stable condition.

Real-world corporate collapses underscore the importance of early detection. For example, when major global retail giants or industrial firms abruptly declare bankruptcy, the warning signs are almost always visible months or years prior within their financial data. When a company's financial health degrades, it typically triggers a cascade of failures across four fundamental business dimensions, or "pillars." To make this analysis accessible to non-financial stakeholders, these pillars can be understood through straightforward personal or medical analogies:

1. Profitability (Net Income & Gross Margin): This measures a company's basic capacity to generate earnings relative to its costs and size. Think of profitability as a body's metabolism, it represents the core energy production of the business. Without positive earnings, a company is effectively starving its operations.
   
3. Liquidity (Current & Cash Ratios): This evaluates whether a business possesses enough immediate cash or easily convertible assets to settle short-term invoices due within the year. Liquidity acts like oxygen; even a highly profitable company can experience sudden structural failure if it runs out of immediate breathing room to pay its daily operational expenses and employee wages.

3. Solvency / Leverage (Debt Ratio & Loan Dependency): This tracks the total volume of long-term debt a company carries relative to its total worth. Solvency represents a firm's skeletal weight burden. While carrying some structural weight (borrowed capital) can help a business grow, carrying excessive debt leaves the company fragile and vulnerable to crashing under market pressure.

4. Efficiency (Asset & Inventory Turnover): This examines how rapidly and effectively a company manages its resources, such as selling out its warehouse stock or utilizing its equipment. Efficiency measures the corporate circulation speed. If inventory sits stagnant on shelves for too long, capital becomes trapped, clogging the company's overall financial health.

By evaluating a dataset containing 6,819 corporate entities, this study addresses a stark operational reality that bankruptcy is a rare yet devastating event. Within this data, only 220 firms succumbed to insolvency (≈3.23%), while 6,599 successfully maintained operations (≈96.77%). This structural imbalance makes manual oversight inefficient. Through data analytics and interactive dashboard design, we can synthesize these complex financial "vital signs" into a clear, unified diagnostic framework allowing investors, creditors, and business managers to identify distress early and intervene before a complete corporate collapse occurs.

## Data Preparation & Filtering

### Data Cleaning

Due to the overwhelming metrics present in the raw data, the analysis process becomes complex. 

So, the first phase of the data cleaning is cutting down the columns into 8 desired columns. Use df.columns.str.strip() to remove hidden spaces in the column headers.

```python
df.columns = df.columns.str.strip()
```

Then, we rename the column headers appropriately using mapping function.

```python
mapping = {
    'Bankrupt?': 'Company Status',
    'Net Income to Total Assets': 'Profit Rate',
    'Operating Gross Margin': 'Gross Margin',
    'Current Ratio': 'Liquidity Ratio',
    'Cash/Total Assets': 'Cash Percentage',
    'Debt ratio %': 'Debt Percentage',
    'Borrowing dependency': 'Loan Dependency',
    'Total Asset Turnover': 'Sales Efficiency',
}
```

Then, clean the hidden spaces from column headers 

```python
cleaned_df = df[list(mapping.keys())].rename(columns=mapping)
```

```python
cleaned_df
```

Now, clean up the long continous value by rounding them up to two decimal places

```python
cleaned_df['Profit Rate'] = (cleaned_df['Profit Rate']).round(4)
cleaned_df['Gross Margin'] = (cleaned_df['Gross Margin']).round(4)
cleaned_df['Liquidity Ratio'] = (cleaned_df['Liquidity Ratio']).round(4)
cleaned_df['Cash Percentage'] = (cleaned_df['Cash Percentage']).round(4)
cleaned_df['Debt Percentage'] = (cleaned_df['Debt Percentage']).round(4)
cleaned_df['Loan Dependency'] = (cleaned_df['Loan Dependency']).round(4)
cleaned_df['Sales Efficiency'] = (cleaned_df['Sales Efficiency']).round(4)
```

```python
cleaned_df
```

Then, replace the 0 and 1 with "Stable" and "Bankrupt" respectively for cleaner look

```python
cleaned_df['Company Status'] = cleaned_df['Company Status'].replace(1, "Bankrupt")
cleaned_df['Company Status'] = cleaned_df['Company Status'].replace(0, "Stable")
```

## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load data (use your actual dataframe name, e.g., cleaned_df)
cleaned_df = pd.read_csv('simple_corporate_data.csv')

# Select key metrics to plot
metrics = ['Profit Rate', 'Debt Percentage', 'Cash Percentage', 'Sales Efficiency']

# Setup the grid
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, col in enumerate(metrics):
    # FIXED: Added hue='Company Status' and legend=False to silence the warning
    sns.boxplot(
        data=cleaned_df, 
        x='Company Status', 
        y=col, 
        ax=axes[i], 
        hue='Company Status', 
        palette='Set2', 
        legend=False
    )
    axes[i].set_title(f'Distribution of {col} by Status')
    axes[i].set_xlabel('Company Status')
    axes[i].set_ylabel(col)

plt.tight_layout()
plt.savefig('metrics_by_status_boxplot.png')
```

```python
numeric_df = cleaned_df.drop(columns=['Company Status'])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Financial Metrics', fontsize=14, pad=15)
plt.tight_layout()
plt.savefig('financial_correlation_heatmap.png')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=cleaned_df, 
    x='Profit Rate', 
    y='Debt Percentage', 
    hue='Company Status', 
    palette={'Stable': '#2ecc71', 'Bankrupt': '#e74c3c'}, 
    alpha=0.6
)

plt.title('Profit Rate vs Debt Percentage by Company Status', fontsize=14)
plt.xlabel('Profit Rate')
plt.ylabel('Debt Percentage')
plt.legend(title='Company Status')
plt.tight_layout()
plt.savefig('profit_vs_debt_scatter.png')
```

### The Distribution Boxplot Analysis

The boxplots break down the actual behavior and distribution spread of these metrics between the two corporate groups.

The boxplot for *Profit_Rate* reveals that healthy firms are tightly clustered around a positive, stable average (0.8101). In contrast, bankrupt firms show a much lower average (0.7381) with a significantly wider downward spread, indicating that many failing firms are bleeding capital rapidly.

Other than that, the distribution for *Cash_Percentage* shows that healthy firms maintain a comfortable safety net, holding an average of 12.66% of their assets in pure cash. Bankrupt firms display a severely compressed distribution box, flatlining at an average of just 4.77% cash. They simply do not have the liquid funds required to stay afloat.

### The Correlation Heatmap Analysis

The heatmap identifies which financial metrics have the strongest statistical connection to corporate failure.

*Profit_Rate* has the strongest negative correlation with bankruptcy (r=-0.315). This is followed closely by *Cash_Percentage* (r=-0.100) and *Gross_Margin* (r=-0.100). This proves mathematically that as a company;s profitability and cash reserves rise, its likelihood of bankruptcy drops significantly.

*Debt_Percentage* (r=0.250) and *Loan_Dependency* (r=0.1760) display strong positive correlations with corporate failure. This confirms that a rising debt load is a direct statistical driver toward structural collapse.

### The Risk-Reward Frontier (Profit vs. Debt Scatter Plot)

The scatter plot maps out the ultimate visual boundary of corporate survival by testing *Debt_Percentage* (X-axis) against *Profit_Rate* (Y-axis).

The vast majority of the teal dots(Healthy Firms) occupy the upper-left quadrant of the plor. This represents high profitability coupled with conservative, low debt which is the ideal state for an enduring business.

The red dots however migrate aggresively toward the bottom-right quadrant. This quadrant visualizes a toxic economic combination where skyrocketing debt percentages mathced with collapsing profit rates. Once a company corsses ito this territory, its structural weight outweighs its energy production, inevitably triggering a bankruptcy event.
